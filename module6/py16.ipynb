```python
def exchange_rates(currency, format='full'):    
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    data = response[currency]     
    if format == 'full':    
        return data      
    elif format == 'value':    
        return data['Value']    
```

```python
def currency_name(id_val):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'  
    response = requests.get(url).json()['Valute']    
    for i in response.values():
        if i['ID'] == id_val:
            return i['Name']
    
currency_name('R01700J')
```

```python
def get_actors(url):
    resp = []
    response = requests.get(url)  
    page = BeautifulSoup(response.text, 'html.parser')  
    ac_block = page.find('div', class_ = 'film-crew-block')
    listAc = ac_block.find_all('li')
    for i in listAc:
        resp.append(i.text)
    return resp
    
get_actors('https://www.kinopoisk.ru/film/42326/')
```

```python
def get_smm_index(group_name, token):
    url = 'https://api.vk.com/method/wall.get'  
    params = {  
        'domain': group_name,  
        'filter': 'owner',  
        'count': 10,  
        'offset': 0,  
        'access_token': token,  
        'v': 5.95  
    }  
    response = requests.get(url, params = params).json()
    count = response['response']['count']
    smm_index = 0
    for record in response['response']['items'][:]:  
        if record['text']:  
            smm_index += record['comments']['count'] + record['likes']['count'] + record['reposts']['count']
    return smm_index / count
get_smm_index('vk', token)
```

```python
from bs4 import BeautifulSoup  
import pandas as pd  
import requests  
def get_html(url):
    r = requests.get(url)
    r.encoding = 'utf-8'
    return r.text
    
def all_links(html):
    soup = BeautifulSoup(html, 'lxml')
    parent_tds = soup.find('div', id = 'listing-filters').find('div', class_ = 'ListingItemSequential-module__container').find_parent()
    parent_tds_str = parent_tds.get('class')[0]
    tds = soup.find('div', id = 'listing-filters').find('div', class_ = parent_tds_str).find_all(
    'div', class_ = 'ListingItemSequential-module__container')
    
    links = []
    
    for i in tds:
        a = i.find('a', class_ = 'Link')
        link = a.get('href')
        links.append(link)
   
    return links
    
def dict_specif(html):
    soup = BeautifulSoup(html, 'lxml')
    tds = soup.find('ul', class_ = 'CardInfo').find_all('li')
    list_cat = []
    for i in range(len(tds)):
        cat = tds[i].find_all('span')[0].text
        value = tds[i].find_all('span')[1].text
        list_cat.append({cat: value})
    return list_cat
    
def name_cost_auto(html):
    soup = BeautifulSoup(html, 'lxml')
    tds_cost = soup.find('div', class_ = 'CardSidebarActions__price-caption')
    tds_name = soup.find('div', class_ = 'CardSidebarActions__title')
    cost = tds_cost.find_all('span')[1].text
    name = tds_name.text
    cost = cost.replace('\xa0', '')
    name = name.replace('\xa0', '')
    list_cat = []
    list_cat.append({'Название':name})
    list_cat.append({'Цена':cost})
    return list_cat
    
links = []
for i in range(1, 100):
    url = 'https://auto.ru/moskva/cars/all/?output_type=table&page=' + str(i)
    html = get_html(url)
    links =  links + all_links(html)
    
data = []
for i in links:
    try:
        html = get_html(i)
        specif = dict_specif(html)
        name_cost = name_cost_auto(html)   
    except:
        continue
    data = data + (name_cost + specif)
```
